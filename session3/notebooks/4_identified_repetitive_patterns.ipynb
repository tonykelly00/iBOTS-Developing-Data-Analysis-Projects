### Merge the pupil-related data as separate npy files

```python
path_to_pupil_data = "../steinmetz_data/raw/subject-Lederberg/session-20171209/behavioral/pupil/"
```

#### Pupil area

```python
pupil_area_data = merge_behavior_data_mat_as_npy(path_to_pupil_data, variable_name="pupil_area")
```

#### Pupil $x$ position

```python
pupil_x_data = merge_behavior_data_mat_as_npy(path_to_pupil_data, variable_name="pupil_x")
```

#### Pupil $y$ position

```python
pupil_y_data = merge_behavior_data_mat_as_npy(path_to_pupil_data, variable_name="pupil_y")
```

```python
n_trials, n_timebins = pupil_y_data.shape
```

```python
fig, ax = plt.subplots(figsize=(4, 3), dpi=150)
trial_idx = 29
sc = ax.scatter(pupil_x_data[trial_idx], pupil_y_data[trial_idx], c=np.linspace(0, 2.5, n_timebins), ec='k', lw=.5)
ax.set(xlabel="pupil x", ylabel="pupil y", title=f"Trial {trial_idx+1}")

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('time (s)') 
```
